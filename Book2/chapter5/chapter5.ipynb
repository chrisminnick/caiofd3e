```python
import datetime as dt
# Today's date (according to the computer's clock)
today = dt.date.today()

# Your birthdate (replace with your actual birthdate)
birthdate = dt.date(2000, 12, 31)

# Calculate the difference between today and your birthdate
delta_age = (today - birthdate)

# Calculate your age in days
days_old = delta_age.days

# Calculate years and months from days_old
years_old = days_old // 365

# Calculate remaining months after accounting for years
months = (days_old % 365) // 30

print(f"You are {years_old} years and {months} months old.")
```

**Determine the difference between your time and UTC time**

```python
here_now = dt.datetime.now()

# --- Legacy approach (naïve datetimes; utcnow() is deprecated in Python 3.12+) ---
import datetime as dt
utc_now = dt.datetime.utcnow() # deprecated in Python 3.12+
time_difference = utc_now - here_now
print(f"My time : {here_now:%I:%M:%S %p}")
print(f"UTC time : {utc_now:%I:%M:%S %p}")
print(f"Time difference: {time_difference}")

# --- Python 3.12+ approach (timezone-aware datetimes) ---
local_now = dt.datetime.now().astimezone()
utc_now = dt.datetime.now(dt.UTC)
time_difference = utc_now - local_now
print(f"My time : {local_now:%I:%M:%S %p %Z} ({local_now:%z})")
print(f"UTC time : {utc_now:%I:%M:%S %p %Z} ({utc_now:%z})")
print(f"Time difference: {time_difference}")
```

**Getting the date and time in different time zones**

```python
# import datetime as dt
import datetime as dt
# import timezone helpers from dateutil
from dateutil.tz import gettz

#UTC time right now
utc=dt.datetime.now(gettz('Etc/UTC'))
print(f"{utc:%A %D %I:%M %p %Z}")

#USA Eastern time
eastern=dt.datetime.now(gettz('US/Eastern'))
print(f"{eastern:%A %D %I:%M %p %Z}")

#USA Central time
central=dt.datetime.now(gettz('US/Central'))
print(f"{central:%A %D %I:%M %p %Z}")

#USA Mountain time
mountain=dt.datetime.now(gettz('US/Mountain'))
print(f"{mountain:%A %D %I:%M %p %Z}")

#USA Pacific time
pacific=dt.datetime.now(gettz('US/Pacific'))
print(f"{pacific:%A %D %I:%M %p %Z}")

```
