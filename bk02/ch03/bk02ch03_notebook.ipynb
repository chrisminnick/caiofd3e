**Internationalizing Python**

Note: `locale.setlocale(locale.LC_ALL, "fr_FR")` works on Windows as printed, but on macOS/Linux it commonly raises `unsupported locale setting` unless the fully-qualified locale (`fr_FR.UTF-8`) is installed/generated first (see Batch2-Errors.md, bk02ch03 p.139-140). This is a platform difference, not a bug in the book's code.

```python
import locale# Just some random number to work withsample_amount = 1234.56# Set the locale to French (France)locale.setlocale(locale.LC_ALL, "fr_FR")# Use locale.currency() to show sample amount in locale formatprint(locale.currency(sample_amount, grouping=True))
```
