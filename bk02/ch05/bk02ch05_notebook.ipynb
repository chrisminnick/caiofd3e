**Determining the difference between your time and UTC time (Figure 5-4)**

Note: `dt.UTC` is a Python 3.12+ alias for `dt.timezone.utc` (the alias itself was added in 3.11). This sandbox runs Python 3.10, so `dt.UTC` raises `AttributeError` here even though the code is correct as printed. Verified the identical logic (and expected ~7-hour offset vs. near-zero elapsed time behavior described on p.168) using the equivalent, pre-3.11 spelling `dt.timezone.utc`.

```python
here_now = dt.datetime.now()# --- Legacy approach (naive datetimes; utcnow() is deprecated in Python 3.12+) ---import datetime as dtutc_now = dt.datetime.utcnow() # deprecated in Python 3.12+time_difference = utc_now - here_nowprint(f"My time : {here_now:%I:%M:%S %p}")print(f"UTC time : {utc_now:%I:%M:%S %p}")print(f"Time difference: {time_difference}")# --- Python 3.12+ approach (timezone-aware datetimes) ---local_now = dt.datetime.now().astimezone()utc_now = dt.datetime.now(dt.UTC)time_difference = utc_now - local_nowprint(f"My time : {local_now:%I:%M:%S %p %Z} ({local_now:%z})")print(f"UTC time : {utc_now:%I:%M:%S %p %Z} ({utc_now:%z})")print(f"Time difference: {time_difference}")
```
