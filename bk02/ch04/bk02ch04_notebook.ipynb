**Concatenating strings (no spaces)**

```python
first_name = "Bart"middle_init = "J"last_name = "Simpson"full_name = first_name + middle_init + last_nameprint(full_name)
```

**Concatenating strings with spaces and punctuation**

```python
first_name = "Bart"middle_init = "J"last_name = "Simpson"full_name = first_name + " " + middle_init + ". " + last_nameprint(full_name)
```

**Using an f-string instead of concatenation**

```python
print(f"{first_name} {middle_init}. {last_name}")
```
