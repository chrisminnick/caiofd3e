# Read CSV options

```python
# File uses semicolons instead of commas
df = pd.read_csv('data.csv', sep=';')

# First column should be the row index
df = pd.read_csv('data.csv', index_col=0)

# File has no header row — provide column names manually
df = pd.read_csv('data.csv', header=None,
                 names=['age', 'income', 'label'])

# Read only certain columns
df = pd.read_csv('data.csv', usecols=['age', 'income'])

# Specify what strings represent missing values
df = pd.read_csv('data.csv', na_values=['N/A', 'missing', '-'])

# Only read the first 1,000 rows (useful for previewing large files)
df = pd.read_csv('data.csv', nrows=1000)
```

# Reading an Excel file

```python
import pandas as pd

# Read the first sheet by default
df = pd.read_excel('sales_data.xlsx')

# Read a specific sheet by name
df = pd.read_excel('sales_data.xlsx', sheet_name='Q3')

# Read a specific sheet by position (0-indexed)
df = pd.read_excel('sales_data.xlsx', sheet_name=1)

# Read all sheets — returns a dictionary of DataFrames
all_sheets = pd.read_excel('sales_data.xlsx', sheet_name=None)
print(all_sheets.keys())  # dict_keys(['Q1', 'Q2', 'Q3', 'Q4'])
df_q1 = all_sheets['Q1']
```

# Parsing Excel sheets as needed

```python
xls = pd.ExcelFile('sales_data.xlsx')
df = xls.parse('Q3', index_col=None, na_values=['NA'])
```

# Loading large files in chunks

```python
# Load 10,000 rows at a time, process each chunk separately
chunk_list = []
for chunk in pd.read_csv('huge_file.csv', chunksize=10000):
# Do some processing on each chunk
    filtered = chunk[chunk['value'] >0]
    chunk_list.append(filtered)

# Combine all processed chunks
df = pd.concat(chunk_list)
```
